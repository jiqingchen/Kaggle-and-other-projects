```python
df.drop(['Sample_Name','TenYearRFS', 'TenRFS', 'SurgToBlood3','ProgressYN', 'RecurYN', 'Bas', 'Bmem', 'Bnv', 'CD4mem',
       'CD4nv', 'CD8mem', 'CD8nv', 'Eos', 'Mono', 'Neu', 'NK', 'Treg','T_cell', 'Lymph', 'CD4T', 'CD8T',
       'NLR', 'MLR', 'CD4_to_8', 'CD4mem_to_nv', 'CD8mem_to_nv', 'Tmem_to_nv','RecOrProg'],axis=1,inplace=True)
df.columns
```

```python
df.replace(' ','_',regex=True, inplace=True)
df.head()
```

```python
X = df.drop(['TenDead','TenYearSurv'],axis=1).copy()
X.head()
```

```python
X2 = X.assign(sex = np.where(X['sex']=='male',1,2),
             grade2 = np.where(X['grade2']=='Grade_1_+_2',1,2),
             smkstat2 = np.where(X['smkstat2']=='non-smoker',1,2),
             ImmunoTx = np.where(X['ImmunoTx']=='Without_Immuno',1,2)).copy()
X2.head()
```

```python
y = df[['TenDead','TenYearSurv']].copy()
print(y.head())
y2=y.assign(outcome = np.where(y['TenDead']==1,y['TenYearSurv'],y['TenYearSurv']*(-1)))['outcome'].copy()
print(y2)
```

```python
import xgboost
xgb_full = xgboost.DMatrix(X2, label=y2)

# create a train/test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X2, y2, test_size=0.2, random_state=7)
xgb_train = xgboost.DMatrix(X_train, label=y_train)
xgb_test = xgboost.DMatrix(X_test, label=y_test)
```

## Train XGBoost model

```python
# use validation set to choose # of trees
params = {
    "eta": 0.002,
    "max_depth": 3,
    "objective": "survival:cox",
    "subsample": 0.5
}
model_train = xgboost.train(params, xgb_train, 1242, evals = [(xgb_test, "test")], verbose_eval=1000)
```

```python
# train final model on the full data set
params = {
    "eta": 0.002,
    "max_depth": 3, 
    "objective": "survival:cox",
    "subsample": 0.5
}
model = xgboost.train(params, xgb_full, 1242, evals = [(xgb_full, "test")], verbose_eval=1000)
```

### Check Performance

```python
def c_statistic_harrell(pred, labels):
    total = 0
    matches = 0
    for i in range(len(labels)):
        for j in range(len(labels)):
            if labels[j] > 0 and abs(labels[i]) > labels[j]:
                total += 1
                if pred[j] > pred[i]:
                    matches += 1
    return matches/total

# see how well we can order people by survival
c_statistic_harrell(model_train.predict(xgb_test), y_test)
```
